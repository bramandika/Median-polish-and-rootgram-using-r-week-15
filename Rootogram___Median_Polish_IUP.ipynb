# Rootogram

* The rootogram is used to test a univariate distribution
* The rootogram is used to compare the observed and expected values ​​graphically by plotting a histogram
* We can see the goodness of fit from the model using rootograms
* This graph compares the root of the observed frequency with the probability model (*fitted probability model*) used
* Observed frequencies are shown as bars and used frequencies (theoritical frequencies*) as lines
* By default, the root scale is used to make smaller frequencies more visible

## What is the difference between a rootogram and a histogram?
* The frequency that describes the height of the bar on the rootgram is the root of the frequency on the histogram

## Types of Rootogram

Rootgram dapat disajikan dalam 3 bentuk:
1. Standing : bar height is represented by **root of observation**
2. Hanging : bar height is represented by **root of observation**
3. Suspend : bar height is represented by **difference between observed roots
with the root of expectation**

# Median Polish

* Median polish is a simple method for obtaining patterns from data summarized in a two-way table.
* The bidirectional table pattern usually describes the difference between data values ​​in rows and columns.
* Rows and columns are factors, while data values ​​are responses
* Two-way table consists of 3 components, row factor, column factor and response

## Introduction

[Data Source](https://mgimond.github.io/ES218/Week11a.html)

## Anatomy of a two-way table

Let us describe the following table which displays the infant mortality rate (per 1000 live births) by region and the level of education of the infant's father from 1964 to 1966 (Hoaglin et al.)
![](https://mgimond.github.io/ES218/img/twoway_01.png)

A two-way table is composed of three variables:
The two-way table is composed of three variables:

- A row factor, which in this example has 4 levels: Northeast, North Central, South and West,
- A column factor, which in this example has 5 levels: under 8, 9-11, 12, 13-15, and 16 or greater,
- The response variable, which in this example has 4(rows) x 5(columns) = 20 values.

The tow-way table is an extension of the one-way table where instead of grouping continuous variables based on one categorical factor, we explore variables beyond 2 categories.

We can express the relationship between the response variable , y, and the two factors as:

$y_{ij}=\ \mu + \alpha_i + \beta_j -\epsilon_{ij}$

Where $y_{ij}$ as response variable for row $i$ and column $j$, $\mu$ as overall typical value (hereinafter called common value), $\alpha_i$ as row effect, $\beta_j$ as effect column and $\epsilon_{ij}$ as residual or value left over after all effects are taken into account.
residual value after all effects have been taken into account.

---

The purpose of this analysis is to describe the response variables according to their effects (eg 2 contributing factors: father's education and region) through an iterative process in which the row and column medians are removed from the response variables repeatedly until the row and column medians approach zero.

## Analysis Workflow

### Create a Dataframe

````r
data <- data.frame(row.names = c("NE","NC","S","W"),
                 ed8       = c(25.3,32.1,38.8,25.4), 
                 ed9to11   = c(25.3,29,31,21.1),
                 ed12      = c(18.2,18.8,19.3,20.3),
                 ed13to15  = c(18.3,24.3,15.7,24),
                 ed16      = c(16.3,19,16.8,17.5)
                 )

data
````

At first glance, it appears that the mortality rate is higher for the NC and S regions, and relatively lower for the W and NE regions. But we can be fooled by the presence of outliers in the data.

Then, we will create a side-by-side boxplot to compare the effects between the two categories. Note that we need to create a long version of the table using package tidyr.

````r
#install.packages('tidyr')
#install.packages('dplyr')
````

````r
library(tidyr)
library(dplyr)

df.l <- data %>%
  mutate(Region = as.factor(row.names(.)) )  %>%
  gather(key=Edu, value = Deaths, -Region) %>% 
  mutate(Edu = factor(Edu, levels=names(data)))

# side-by-side plot
OP <- par(mfrow=c(1,2))
plot(Deaths ~ Region + Edu, df.l)
par(OP)
````

At this point, the plots show a possible effect of the length of father's education and regional effects on infant mortality rates with the initial effect being more important.

Next we will attempt to quantify these effects using Tukey's median polish.

Next we will attempt to measure these effects using **Tukey's median polish**

### Median Polish

#### Step 1 : Compute overall median and residual table

First, we will calculate the median value for all data; this will be the initial estimate for the common value. The result of the median value is placed in the upper left corner of the table. Create a residual table containing the reduction of the original value with the overall median.

![](https://mgimond.github.io/ES218/img/twoway_02.jpg)

We also note that the row and column effects are filled as zeros.

#### Step 2: Compute the row medians

Next, calculate the median of rows (as in the figure) from the table of residuals.

![](https://mgimond.github.io/ES218/img/twoway_03.jpg)

Note that in addition to calculating the row median of the response variables, we also calculated the median value of the column effect in the top margin (which is currently zero because we don't have a column effect yet).

#### Step 3: Create a new residual table from the row medians

Add the median of the rows to the left margin of the residual table (as shown in the figure). This value represents the row effect in the first iteration. Create a new residual value by subtracting the value of each cell by the row median of each cell.

![](https://mgimond.github.io/ES218/img/twoway_04.jpg)

Subtractions are also made for column effects (even though they are currently zero) and global medians (common effects).

#### Step 4 : Compute the column medians

Next, calculate the median of the column (as shown) from the table of residuals. Note that we are NOT calculating the column median from the original value of the response variable. Also note we calculated the median value of the row effect as well (left margin).

![](https://mgimond.github.io/ES218/img/twoway_05.jpg)

#### Step 5 : Create a new residual table from the column medians

Create a new residual table with the value of each cell in the form of subtraction of each previous residual value with the median of each column. For example, for the cell in the upper left corner, the residual value is $7.0−7.4=−0.4$.

Column medians are also added to the column effect margin above the table. Note that we also added the line effect median, `-0.5` to the common effect.

![](https://mgimond.github.io/ES218/img/twoway_06.jpg)

We have now completed the first row and column iteration. However, there may still be row and column effects that can be drawn from the residuals. Therefore we move on to the second iteration.

#### Step 6: Second iteration – row effects

Next, we calculate the row median of the residuals, then add the column effect median to the top margin (and common value) and subtract the residual value from the median value. Like step 3.

![](https://mgimond.github.io/ES218/img/twoway_06b.jpg)

#### Step 7 : Second iteration – column effects

To close the second iteration, we calculate the column median of the residuals and then subtract the residual value from the column median. We also add the median value to the row effect and the common value value.

![](https://mgimond.github.io/ES218/img/twoway_07.jpg)

#### When do we stop iterating?

The end goal is to iterate over row and column operations until the median of column and row effects is close to 0. However, Hoaglin et al. (1983) caution against "using controlless iteration" and suggest that multiple steps should often be more than sufficient. In this example, the third iteration might just be the guarantee. The process is as follows.

![](https://mgimond.github.io/ES218/img/twoway_08.jpg)

The final form of the table is as follows:

![](https://mgimond.github.io/ES218/img/twoway_09.jpg)

The progress for each iteration is shown side by side in the following figure. It can be seen how at the end of the iteration, the common factor contributes as the biggest effect. It is also worth looking at a very high residual in the Southern region associated with grade 8 education levels.

![](https://mgimond.github.io/ES218/Week11a_files/figure-html/clock-1.png)

#### Interpreting the median polish

As noted earlier, the two-way table represents the relationship between the response variable y and the two categories as:

$y_{ij}=\ \mu + \alpha_i + \beta_j -\epsilon_{ij}$

In the example above, $\mu$ = 20.6; $\alpha_i$ = -1.5, 2.6 -0.4, and 0.4 for $i$ = `NE`, `Nc`, `S` and `W` respectively; $\beta_j$ = 7.6, 6.0, -0.9, 0.2 and -3.5 for $j$ = `ed8`, `ed9to11`, `ed12`, `ed13to15` and `ed16` respectively.

The residual value represents a measure of the mortality rate that cannot be explained by both factors.

So that the death rate in the top left corner of the cell from the table can be reshaped to be.
![](https://mgimond.github.io/ES218/img/twoway_10.jpg)

Examination of the table shows that
- the highest infant mortality rate for fathers who have not attended school for more than 8 years, indicated by a high column effect of 7.6.
- This is the infant mortality rate relative to the median overall (ie on average, 20.6 infants per thousand die each year and the rate continues to rise by 7.6 + 20.6 for infants whose fathers did not pass 8th grade).
- Infants whose fathers have studied more than 16 years have a lower mortality rate, indicated by a lower effect value of -3.5.
- The regional effect also shows high mortality rates in the North Central and Western regions (with effect values ​​of 2.55 and 0.35) and low mortality rates in the North Eastern and Southern regions; but the regional effect is less dominant than the old effect of father's education.

It is also important to look at the distribution of the residual values ​​in the two-way table. For one thing we need to identify abnormally high or low residuals in the table (eg residuals in the southern region that need further exploration)).

## Implementing the median polish in R

The steps highlighted in the previous section can easily be implemented in writing or using a spreadsheet for larger data sets. R has a built-in function named `medpolish()` to do this process. We can define the number of iterations by setting the parameter `maxiter=` but note that `medpolish` will by default estimate the best value for the number of iterations. We try to use the value 3 to equate the result with the calculation of the previous 3 iterations.

````r
data.med <- medpolish(data, maxiter=3)
````

The three values ​​displayed in the console are the absolute sum values ​​of the residuals in each iteration; we will ignore this value. The output of the model is stored in the data.med object. To view the contents, just type in the command line.

````r
data.med
````


Three effects are shown along with the residuals (it should be noted that the precision shown is higher than the one we used in the previous analysis).

## What if we use the mean instead of the median?

The procedure is the same with a few differences.

We calculate the global mean instead of the median (common effect) and then subtract it from each value in the table. Then we calculate the row mean (row effect) and then subtract it from each value in the corresponding row. Finally we calculate the column mean (of the residuals) as the column effect. Just that.

Unlike median polish, we do not iterate over the smoothing process. An example of the "mean" polish can be seen as follows:

![](https://mgimond.github.io/ES218/img/twoway_11.png)

The results differ slightly from those using the median polish. Given that the mean is sensitive to outliers and the median is not. If a robust estimate of the effect is required, stick to the median polish.

So what can we get from the "mean" polish? In fact, it is the same as the basis of a two-way ANOVA.
